```python
import numpy as np
import pandas as pd
```

```python
#исходные данные x
data = {
    'x0':[1 for i in range(10)],
    'x1':[2,13,6,10,6,7,4,11,9,9],
    'x2':[4,7,1,7,11,14,8,15,5,7],
    'x3':[8,2,3,3,14,10,5,11,8,4],
}

#нечеткие y
data_output = {
    'm':[10,12,5,10,19,20,11,22,11,13],
    'l':[4,5,3,5,8,6,4,8,6,4],
    'r':[4,5,3,5,8,6,4,8,6,4],
}
```

```python
df_x = pd.DataFrame(data)
df_y = pd.DataFrame(data_output)
```

```python
X = df_x.values
interim_val = np.linalg.inv(X.T @ X) @ X.T

c_t = interim_val @ df_y['m'].values
a_t = interim_val @ df_y['l'].values
b_t = interim_val @ df_y['r'].values
```

```python
#параметры для уравнения
param_estimate = np.array([c_t,a_t,b_t])
```

```python
y_m = [(c_t * vector_x).sum() for vector_x in X]
y_l = [(a_t * vector_x).sum() for vector_x in X]
y_r = [(b_t * vector_x).sum() for vector_x in X]
table = {
    'y_m': y_m,
    'y_l': y_l,
    'y_r': y_r,
    'm_error':y_m - df_y.m,
    'l_error':y_l - df_y.l,
    'r_error':y_r - df_y.r,
    'm_error_accuracy': 1 - np.abs((y_m - df_y.m)/df_y.m),
    'l_error_accuracy': 1 - np.abs((y_l - df_y.l)/df_y.l),
    'r_error_accuracy': 1 - np.abs((y_r - df_y.r)/df_y.r),
}

final_table = pd.DataFrame(table)
```
